```python
%sql
CREATE TABLE funnel (
    user_id INT,
    step STRING
);
```

```python
%sql
INSERT INTO funnel VALUES
(1, 'view'),
(1, 'cart'),
(2, 'view'),
(2, 'cart'),
(1, 'purchase');
```

```python
%sql
--Count users who completed all steps (view -> cart -> purchase)
SELECT COUNT(*) AS completed_users
FROM (
    SELECT user_id
    FROM funnel
    GROUP BY user_id
    HAVING COUNT(DISTINCT step) = 3
);
```
