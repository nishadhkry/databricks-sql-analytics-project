```python
%sql
CREATE TABLE transactions (
    user_id INT,
    txn_date DATE,
    amount INT
);
```

```python
%sql
INSERT INTO transactions VALUES
(1, '2024-01-01', 100),
(1, '2024-02-01', 200),
(2, '2024-01-05', 300);
```

```python
%sql
--latest transaction for each user
SELECT user_id, txn_date, amount
FROM (
    SELECT *,
           ROW_NUMBER() OVER (
               PARTITION BY user_id
               ORDER BY txn_date DESC
           ) AS rn
    FROM transactions
) t
WHERE rn = 1;
```
