```python
%sql
CREATE TABLE user_activity (
    user_id INT,
    activity_date DATE
);
```

```python
%sql
INSERT OVERWRITE user_activity VALUES
(1, '2024-01-01'),
(2, '2024-01-01'),
(1, '2024-01-02'),
(3, '2024-01-02'),
(2, '2024-01-02');
```

```python
%sql
-- number of active users per day
SELECT 
    activity_date,
    COUNT(DISTINCT user_id) AS daily_active_users
FROM user_activity
GROUP BY activity_date
ORDER BY activity_date;
```

```python
%sql
-- Find users who logged in on consecutive days
SELECT DISTINCT a.user_id
FROM user_activity a
JOIN user_activity b
ON a.user_id = b.user_id
AND b.activity_date = DATE_ADD(a.activity_date, 1);
```
