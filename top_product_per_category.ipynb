```python
%sql
-- Create Table
CREATE  TABLE IF NOT EXISTS sales (
    category STRING,
    product STRING,
    revenue INT
);
```

```python
%sql
INSERT OVERWRITE sales VALUES
('Phone', 'iPhone', 5000),
('Phone', 'Samsung', 4200),
('Laptop', 'Dell', 7000),
('Laptop', 'HP', 6500),
('Tablet', 'iPad', 3000),
('Tablet', 'Samsung Tab', 2800);
```

```python
%sql
-- Query
SELECT category, product, revenue
FROM (
    SELECT category,
           product,
           revenue,
           ROW_NUMBER() OVER (
               PARTITION BY category
               ORDER BY revenue DESC
           ) AS rn
    FROM sales
) t
WHERE rn = 1;
```
